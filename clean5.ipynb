# 处理开盘没有成交的股票


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import Ipynb_importer
from utils_x import *
from parameters_x import *
import seaborn as sns
import datetime
from scipy.stats import norm
sns.set(style="darkgrid")
import logging
import time
```

```python
begin_interval_df = pd.read_csv('begin_930_interval_df.csv', index_col=0)
begin_interval_df["date_time"] = pd.to_datetime(begin_interval_df["date_time"])
```

```python
begin_interval_df["open_mid"] = begin_interval_df.apply(lambda x: np.nan if x["type"]==-100 else x["open_mid"], axis=1)
```

```python
begin_interval_df["close_mid"] = begin_interval_df.apply(lambda x: np.nan if x["type"]==-100 else x["close_mid"], axis=1)
```

```python
gb = begin_interval_df.groupby(["unique_symbol"])
begin_interval_df["t+1_open_mid"] = gb["open_mid"].transform(lambda x: x.shift(-1))
begin_interval_df["t+1_close_mid"] = gb["close_mid"].transform(lambda x: x.shift(-1))
```

```python
begin_interval_df.to_csv("begin_930_interval_df.csv")
```
