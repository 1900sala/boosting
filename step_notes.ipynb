```python
import numpy as np
import pandas as pd
import os
```

```python
path = "/home/huandong/boosting/log"
```

```python
def func(path):
    path_list = []
    for i in os.listdir(path):
        path_list.append(os.path.join(path,i))
    return path_list
```

```python
log_file = func(path)
df_dic = {
          "factor":[],
          "train_set2010":[],
          "train_set2011":[],
          "train_set2012":[],
          "train_set2013":[], 
          "train_set2014":[],
          "train_set2014":[],
          "train_corrcoef":[],
          "train_corrcoef_Nodemean":[],
          "val_set2014":[],
          "val_set2015":[],
          "val__all_corrcoef":[],
          "val__all_corrcoef_Nodemean":[],
          "test_set2015":[],
          "test_set2016":[],
          "test_set2017":[],
          "test_all_corrcoef":[],
          "test_all_corrcoef_Nodemean":[],
          "return auto corr":[],
          "tree_nums":[],
           "pred auto corr":[],
           "params":[]
          }
```

```python
for f in log_file:
    if f[-3:] == "txt":
        file = open(f, "r")
        rlist = file.readlines()
    else:
        continue
    if len(rlist) != 27:
        continue
    for r in rlist:
        if r == "":
            continue
        temp = r.split(' - INFO - ')[1]
        if temp[0] == "{":
            df_dic["params"].append(temp[1:-2])
        temp_list = temp.split(":")
        if len(temp_list) == 2:
            if temp_list[0]=="factor":
                df_dic[temp_list[0]].append(temp_list[1][:-1])           
            else:
                df_dic[temp_list[0]].append(np.float(temp_list[1][:-1]))
            
```

```python
df = pd.DataFrame(df_dic)
```

```python
df = df.sort_values("factor")
```

```python
df["level"] = df["factor"].apply(lambda x: len(x.split("|")))
df["factor"] = df["factor"].apply(lambda x: x.replace("row", "raw"))
```

```python
df["type_raw"] = df["factor"].apply(lambda x: 0 if len(x.split("raw")) >1 else 1)
```

```python
df["type_ema"] = df["factor"].apply(lambda x: 0 if len(x.split("e")) >1 else 1)
df["type_IC"] = df["factor"].apply(lambda x: 0 if len(x.split("IC")) >1 else 1)
```

```python
df["params"] = df["params"].apply( lambda x: x.split(","))
df["params"] = df["params"].apply( lambda x: [i.split(":") for i in x])
params_dic = {}
dic_key = ['max_depth','subsample','colsample_bytree','alpha','nthread','num_parallel_tree','tree_method','min_child_weight']
for key in dic_key: params_dic[key] = []
for _, row in  df.iterrows():
    for i in row["params"]:
        i[0] = i[0].replace(' ', '')
        i[0] = i[0].replace("'", '')
        i[1] = i[1].replace(' ', '')
        i[1] = i[1].replace("'", '')
        if i[0] in dic_key:
            params_dic[i[0]].append(i[1])
    max_len = len(params_dic["max_depth"])
    for key in dic_key:
        if len(params_dic[key])<max_len:
            params_dic[key].append("")
for key in dic_key: df[key] = params_dic[key]
```

```python
df.to_csv("step_notes.csv")
```
