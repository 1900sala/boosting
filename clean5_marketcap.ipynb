```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import Ipynb_importer
import seaborn as sns
import datetime
from scipy.stats import norm
sns.set(style="darkgrid")
import logging
import time
import os
```

```python
def func(path):
    path_list = []
    for i in os.listdir(path):
        path_list.append(i)
    return path_list
```

```python
t1 = datetime.datetime(2010,1,1)
t2 = datetime.datetime(2019,7,1)


df = pd.DataFrame()
path_base = '/data/share/china_stocks/data/factor_data/'
for y in ["2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019"]:
    path = path_base + y
    path_list = func(path)
    for s in range(len(path_list)):
        if s%100 == 0:
            print(y, s, s/len(path_list))
        filename = "barra_asset_price_" + path_list[s] + ".csv"
        temp = pd.read_csv(os.path.join(path, path_list[s], filename))
        temp["date_time"] = path_list[s]
        temp["date_time"] = pd.to_datetime(temp["date_time"])
        df = df.append(temp)
    #     print(temp)
```

```python
unique_symbol = [int(str(i)[-6:]) for i in list(df["tid"])]
df["unique_symbol"] = unique_symbol
```
